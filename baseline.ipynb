```python
import numpy as np
import pandas as pd
from lightgbm import LGBMClassifier
import warnings
warnings.filterwarnings('ignore')
```

```python
train  = pd.read_csv('/content/drive/MyDrive/공동AI경진대회/예선/competition_data/train.csv')
test  = pd.read_csv('/content/drive/MyDrive/공동AI경진대회/예선/competition_data/test.csv')
```

```python
train = train.drop(['index', 'country'],axis = 1)
test =test.drop(['index', 'country'],axis = 1)
```

```python
#train을 target과 feature로 나눠줍니다.
train_x=train.drop(['nerdiness'], axis=1)
train_y=train['nerdiness']
```

```python
lgbm_clf = LGBMClassifier(
            n_estimators=1000
        )
lgbm_clf.fit(train_x, train_y)

lgbm_pred = lgbm_clf.predict(test)
```

```python
submission = pd.read_csv('/content/drive/MyDrive/공동AI경진대회/예선/competition_data/sample_submission.csv')

submission
```

```python
submission["nerdiness"] = lgbm_pred
```
